```python
platelet_data = np.array([169.56519478, 167.52997358, 162.25797711, 166.17642794,
       164.30340658, 165.01533004, 165.23749839, 164.10596845,
       163.20133229, 163.37756603, 161.54313575, 167.70804782,
       166.98543108, 164.06967058, 160.18714044, 163.39025427])
```

```python
platelet_data_norm = (platelet_data - np.mean(platelet_data))/(np.std(platelet_data))

sm.qqplot(platelet_data_norm,line='45')
```

**Perform t-test**

```python
#Calculate the test statistic
Tn = abs(np.mean(platelet_data) - 270)/np.std(platelet_data)
print(Tn)
```

```python
#Calculate the q-alpha value at alpha=5% is 25 at degree of freedom = 14
#Tool used to calculate q-alpha at 5%: https://shiny.rit.albany.edu/stat/chisq/
q_alpha = 25
psi_alpha = (Tn > q_alpha)
print(psi_alpha)
```

**We started with the null hypothesis of no thrombocytopenia
Conclusion: We can reject the null hypothesis that of negative thrombocytopenia ****
